```python
# df1 sort for different genders

df1_f = df1[df1["Sex"] == 'female']
df1_m = df1[df1["Sex"] == 'male']

# survived for different genders
df1_f_S = df1_f[df1_f['Survived'] == 1]
df1_m_S = df1_m[df1_m['Survived'] == 1]


# female sort for display
df1_f
```

```python
#Calculation of survival rate for different Genders
survival_rate_f = len(df1_f_S) / len(df1_f)
survival_rate_m = len(df1_m_S) / len(df1_m) 

print(survival_rate_f)
print(survival_rate_m)
```

```python
# df1 sort for different ages (e.g. 'minors', 'adults', 'seniors' )
# note: for some passangers age is NaN, taken into account
df1_min = df1[df1["Age"] < 18]
df1_ad = df1[(df1["Age"] > 18)&(df1["Age"] < 60)]
df1_sen = df1[df1["Age"] < 60]
df1_age_na = df1[df1["Age"].isna() == True]

# survived for different ages
df1_min_S = df1_min[df1_min['Survived'] == 1]
df1_ad_S = df1_ad[df1_ad['Survived'] == 1]
df1_sen_S = df1_sen[df1_sen['Survived'] == 1]
df1_age_na_S = df1_age_na[df1_age_na['Survived'] == 1]

#df1_min for display
df1_min
```

```python
#Calculation of survival rate for different ages
survival_rate_min = len(df1_min_S) / len(df1_min)
survival_rate_ad = len(df1_ad_S) / len(df1_ad)
survival_rate_sen = len(df1_sen_S) / len(df1_sen)
survival_rate_age_na = len(df1_age_na_S) / len(df1_age_na) 


print(survival_rate_min)
print(survival_rate_ad)
print(survival_rate_sen)
print(survival_rate_age_na)
```

```python
# df1 sort for Siblings and spouses onboard
df1_SibSp_0 = df1[df1["SibSp"] == 0]
df1_SibSp_1plus = df1[df1["SibSp"] > 0]

# survived for different SibSp

df1_SibSp_0_S = df1_SibSp_0[df1_SibSp_0['Survived'] == 1]
df1_SibSp_1plus_S = df1_SibSp_1plus[df1_SibSp_1plus['Survived'] == 1]

# SibSp 0 for display
df1_SibSp_0
```

```python
#Calculation of survival rate for different SibSp
survival_rate_SibSp_0 = len(df1_SibSp_0_S) / len(df1_SibSp_0)
survival_rate_SibSp_1plus = len(df1_SibSp_1plus_S) / len(df1_SibSp_1plus) 

print(survival_rate_SibSp_0)
print(survival_rate_SibSp_1plus)
```

```python
# df1 sort for parents and children onboard
df1_Parch_0 = df1[df1["Parch"] == 0]
df1_Parch_1plus = df1[df1["Parch"] > 0]

# survived for different Parch

df1_Parch_0_S = df1_Parch_0[df1_Parch_0['Survived'] == 1]
df1_Parch_1plus_S = df1_Parch_1plus[df1_Parch_1plus['Survived'] == 1]

# Parch 0 for display
df1_Parch_0
```

```python
#Calculation of survival rate for different Parch
survival_rate_Parch_0 = len(df1_Parch_0_S) / len(df1_Parch_0)
survival_rate_Parch_1plus = len(df1_Parch_1plus_S) / len(df1_Parch_1plus) 

print(survival_rate_Parch_0)
print(survival_rate_Parch_1plus)
```

```python
# df1 sort for different Embarked values
df1_Embarked_C = df1[df1["Embarked"] == "C"]
df1_Embarked_S = df1[df1["Embarked"] == "S"]
df1_Embarked_Q = df1[df1["Embarked"] == "Q"]
# survived for different Embarked values

df1_Embarked_C_S = df1_Embarked_C[df1_Embarked_C['Survived'] == 1]
df1_Embarked_S_S = df1_Embarked_S[df1_Embarked_S['Survived'] == 1]
df1_Embarked_Q_S = df1_Embarked_Q[df1_Embarked_Q['Survived'] == 1]

# Embarked_C for display
df1_Embarked_C
```

```python
#Calculation of survival rate for different Embarked values
survival_rate_Embarked_C = len(df1_Embarked_C_S) / len(df1_Embarked_C)
survival_rate_Embarked_S = len(df1_Embarked_S_S) / len(df1_Embarked_S) 
survival_rate_Embarked_Q = len(df1_Embarked_Q_S) / len(df1_Embarked_Q)

print(survival_rate_Embarked_C)
print(survival_rate_Embarked_S)
print(survival_rate_Embarked_Q)
```

```python
#defining functions required to compute the probability

def def_Pclass(row):
    if row['Pclass'] == 3:
        return survival_rate_Pclass_3
    elif row['Pclass'] == 2:
        return survival_rate_Pclass_2
    else:
        return survival_rate_Pclass_1

def def_Sex(row):
    if row['Sex'] == 'male':
        return survival_rate_m
    elif row['Sex'] == 'female':
        return survival_rate_f

def def_SibSp(row):
    if row['SibSp'] == 0:
        return survival_rate_SibSp_0
    else:
        return survival_rate_SibSp_1plus

def def_Parch(row):
    if row['Parch'] == 0:
        return survival_rate_Parch_0
    else:
        return survival_rate_Parch_1plus

def def_Embarked(row):
    if row['Embarked'] == 'Q':
        return survival_rate_Embarked_Q
    elif row['Embarked'] == 'S':
        return survival_rate_Embarked_S
    else:
        return survival_rate_Embarked_C

def calculate_survival_probability(df):
    
    df['SurvivalProbability'] = 100

    
    df['SurvivalProbability'] *= df.apply(def_Pclass, axis=1)
    df['SurvivalProbability'] *= df.apply(def_Sex, axis=1)
    df['SurvivalProbability'] *= df.apply(def_SibSp, axis=1)
    df['SurvivalProbability'] *= df.apply(def_Parch, axis=1)
    df['SurvivalProbability'] *= df.apply(def_Embarked, axis=1)

    
    df['Survived'] = (df['SurvivalProbability'] >= 0.5).astype(int)
    del df['SurvivalProbability']

    return df
```
